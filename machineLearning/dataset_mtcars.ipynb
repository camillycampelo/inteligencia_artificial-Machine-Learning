## 1. Qual é a relação entre o número de cilindros (cyl) e o consumo de combustível em milhas por galão (mpg)?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
!pip install pydataset
from pydataset import data

mtcars = data('mtcars')
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='cyl', y='mpg', data=mtcars, palette='Set2')
plt.title('Relação entre Número de Cilindros e Consumo de Combustível (mpg)', fontsize=14)
plt.xlabel('Número de Cilindros (cyl)', fontsize=12)
plt.ylabel('Consumo de Combustível (mpg)', fontsize=12)
plt.show()
```

# 2. O tipo de transmissão (am - automática ou manual) influencia a potência do motor (hp)?

```python
hp_by_transmission = mtcars.groupby('am')['hp'].mean()

plt.figure(figsize=(8, 8))
plt.pie(hp_by_transmission, labels=['Automática', 'Manual'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Distribuição da Potência do Motor (hp) por Tipo de Transmissão', fontsize=14)
plt.show()
```

# 3. Existe correlação entre o peso do carro (wt) e o tempo de aceleração de 1/4 de milha (qsec)?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='wt', y='qsec', data=mtcars, color='blue', s=100, edgecolor='w')
plt.title('Correlação entre Peso do Carro (wt) e Tempo de Aceleração (qsec)', fontsize=14)
plt.xlabel('Peso do Carro (wt)', fontsize=12)
plt.ylabel('Tempo de Aceleração (qsec)', fontsize=12)
plt.show()
```

# 4. Carros com mais carburadores (carb) tendem a ter maior cilindrada do motor (disp)?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='carb', y='disp', data=mtcars, palette='Set1')
plt.title('Relação entre Número de Carburadores (carb) e Cilindrada (disp)', fontsize=14)
plt.xlabel('Número de Carburadores (carb)', fontsize=12)
plt.ylabel('Cilindrada (disp)', fontsize=12)
plt.show()
```

# 5. Qual é a média de milhas por galão (mpg) para carros com diferentes números de marchas (gear)?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='gear', y='mpg', data=mtcars, palette='Set2', ci=None)
plt.title('Média de Consumo de Combustível (mpg) por Número de Marchas (gear)', fontsize=14)
plt.xlabel('Número de Marchas (gear)', fontsize=12)
plt.ylabel('Consumo Médio de Combustível (mpg)', fontsize=12)
plt.show()
```
