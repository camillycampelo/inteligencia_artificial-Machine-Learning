# Random Forest



Técnica de aprendizado de máquina que combina várias árvores de decisão. Utilizado para classificação de dados.

Passo a passo:

*   Importar as bilbiotecas
*   Carregar o conjunto de dados
*   Separar as características
*   Dividir em dados de teste e treinamento
*   Treinar o modelo
*   Fazer uma previsão com os dados de teste
*   Avaliar a precisão do modelo
*   Imprimir o resultado da precisão

Depois de aprovado o modelo:

* Realizar a previsão com novos dados



```python
#Bibliotecas
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
dados = pd.read_csv('/content/Planilha sem título - Página1.csv')
x = dados.drop('faz_emprestimo', axis=1) # Sem o alvo
y = dados['faz_emprestimo'] # Alvo

```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
modelo = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
modelo
```

```python
modelo.fit(x_treino, y_treino)
```

```python
previsao = modelo.predict(x_teste)
print("O cliente fará empréstimo?", previsao)
```

```python
acuracia = accuracy_score(y_teste, previsao)
print("Acurácia:", acuracia)
```

```python
novos_dados = pd.DataFrame({
    'idade':[22],
    'renda':[2800],
    'emprestimo':[1]
})

previsao_novos_dados = modelo.predict(novos_dados)
print("Previsão com novos dados", previsao_novos_dados)
```

```python
import matplotlib.pyplot as plt

importancia_caracteristicas = modelo.feature_importances_
caracteristicas = x.columns

plt.bar(caracteristicas, importancia_caracteristicas)
plt.xlabel('Características')
plt.ylabel('Importância')
plt.title('Importância das Características')
```
